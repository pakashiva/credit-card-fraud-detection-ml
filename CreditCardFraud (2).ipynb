# 2. Feature Engineering

Since most of the columns except 'Time' and 'Amount' are prescaled , we will only focus on scaling the remaining features.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()

df = data.copy()

df['Amount_scaled'] = scalar.fit_transform(df[['Amount']])
df['Time_Scaled'] = scalar.fit_transform(df[['Time']])

df = df.drop(columns=['Amount' , 'Time'])

df.head(3)
```

# 3. Model Comparision

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report , roc_auc_score , confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay , roc_curve , precision_recall_curve , average_precision_score, auc
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
X = df.drop(columns=['Class'])
y = df['Class']

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.2,  stratify=y , random_state=94)
```

```python
models = {
    'Logistic' : LogisticRegression(class_weight='balanced' , max_iter=1000) , 
    'Random forest' : RandomForestClassifier(class_weight='balanced'),
    'XGB classifier' : XGBClassifier(class_weight= 'balanced' , eval_metric='logloss')
}
```

```python
for model_name , model in models.items():

    print(f"Model: {model_name}")
    print("="*50)
    
    model.fit(X_train , y_train)
    y_pred = model.predict(X_test)

    y_proba = model.predict_proba(X_test)[: , 1]

    print('The ROC-AUC score: ' , roc_auc_score(y_test , y_proba))
    print('\n')
    print(classification_report(y_test , y_pred))
    print('\n')
    pr_auc = average_precision_score(y_test, y_proba)
    print("PR-AUC:", pr_auc)
    print('\n')
    cm = confusion_matrix(y_test , y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title('Confusion Matrix')
    plt.ylabel('Actual values')
    plt.xlabel('Predicted values')
    plt.show()
```

Three baseline models — Logistic Regression, Random Forest, and XGBoost — were evaluated using ROC-AUC, Precision, Recall, and F1-score. Although XGBoost achieved a slightly higher ROC-AUC, Random Forest produced the highest F1-score (0.83) for the fraud class while maintaining very high precision (0.96) and the same recall (0.74) as XGBoost. In fraud detection problems with highly imbalanced data, metrics such as F1-score and Precision-Recall performance are more informative than accuracy or ROC-AUC. Random Forest provided a better balance between detecting fraudulent transactions and minimizing false positives. Therefore, Random Forest was selected as the baseline model for further tuning and imbalance-handling techniques.

# Sampling Techniques:

```python
# Function to quantify a technique

def quantify_technique(sampler , X_train , y_train , X_test , y_test):
    X_train_res , y_train_res = sampler.fit_resample(X_train , y_train)

    model = RandomForestClassifier(random_state=94)
    model.fit(X_train_res , y_train_res)

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[: , 1]

    report = classification_report(y_test , y_pred)
    roc_auc = roc_auc_score(y_test , y_proba)
    cm = confusion_matrix(y_test , y_pred)
    pr_area = average_precision_score(y_test , y_proba)

    print(f'ROC-AUC score: {roc_auc}')
    print('--'*25)
    print(f'Precision-Recall-AUC: {pr_area}')
    disp = ConfusionMatrixDisplay(confusion_matrix = cm)
    disp.plot()
    plt.title('Confusion Matrix')
    
    labels = ["Not Fraud", "Fraud"]
    plt.xticks(np.arange(len(labels)), labels)
    plt.yticks(np.arange(len(labels)), labels)
    
    plt.ylabel('Actual values')
    plt.xlabel('Predicted values')
    plt.show()
    
    print(report)
    
```

## 1. UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
sampler = RandomUnderSampler(random_state=94)
quantify_technique(sampler , X_train , y_train , X_test , y_test)
```

## 2. OverSampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
over_sampler = RandomOverSampler(random_state=94)

quantify_technique(over_sampler , X_train , y_train , X_test , y_test)
```

## 3. Synthetic Minority Oversampling Technique(SMOTE)

```python
from imblearn.over_sampling import SMOTE
```

```python
smote_sampler = SMOTE(random_state=94)

quantify_technique(smote_sampler , X_train , y_train , X_test , y_test)
```

## 4. SMOTE-TOMEK LINKS

```python
from imblearn.combine import SMOTETomek
```

```python
smtomel = SMOTETomek(random_state=94)

quantify_technique(smtomel ,  X_train , y_train , X_test , y_test)
```

```python
print(f'Cross validation score(f1-score): {np.round(np.mean(score) , 2)}')
```

```python
# Function to plot confusion matrix
def plotConfusionMatrix(cm):
    disp = ConfusionMatrixDisplay(confusion_matrix = cm)
    disp.plot()
    plt.title('Confusion Matrix')
        
    labels = ["Not Fraud", "Fraud"]
    plt.xticks(np.arange(len(labels)), labels)
    plt.yticks(np.arange(len(labels)), labels)
        
    plt.ylabel('Actual values')
    plt.xlabel('Predicted values')
    plt.show()
```

# Hyperparameter Tuning

```python
from imblearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
pipe = Pipeline([
    ('smote' , SMOTE(random_state=94)) , 
    ('model' , RandomForestClassifier(random_state=94 , n_jobs=1))
])

parameter_grid = {
    'model__n_estimators' : [100, 300],
    'model__max_depth' : [10 , 15],
    'model__min_samples_split' : [10 , 20],
    'model__min_samples_leaf' : [5, 10]
}
```

```python
grid = GridSearchCV(
    estimator=pipe,
    param_grid=parameter_grid,
    cv = sfk,
    scoring='f1',
    n_jobs=-1
)

grid.fit(X_train , y_train)
```

```python
best_model = grid.best_estimator_
best_params = grid.best_params_

print('The best parameters of the modes:')
print('--'*25)
print(best_params)

y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[: , 1]

report = classification_report(y_test , y_pred)
roc_auc = roc_auc_score(y_test , y_proba)
cm = confusion_matrix(y_test , y_pred)
pr_area = average_precision_score(y_test , y_proba)

print("Best CV Score:", grid.best_score_)
print('--'*25)
print(f'ROC-AUC score: {roc_auc}')
print('--'*25)
print(f'Precision-Recall-AUC: {pr_area}')

plotConfusionMatrix(cm)
    
print(report)
```

To address class imbalance, both resampling techniques (SMOTE, SMOTE-Tomek) and algorithm-level methods (class_weight in Random Forest) were evaluated. Although SMOTE improved the model compared to simple undersampling and oversampling, the Random Forest model using class-weight balancing achieved slightly better performance after hyperparameter tuning.

This approach adjusts the model’s learning process by assigning higher importance to the minority class during training, allowing the model to better detect fraudulent transactions without generating synthetic samples. Additionally, avoiding synthetic data helps preserve the original distribution of the dataset and reduces the risk of introducing noise or overfitting.

Therefore, the Random Forest model with class_weight='balanced' was selected as the final approach, as it provided the best trade-off between precision, recall, and F1-score for the fraud class.

# Cost Sensitive Learning

```python
final_model = RandomForestClassifier(
            n_estimators=100,
            max_depth=15, 
            min_samples_leaf= 5,
            min_samples_split= 10, 
            n_jobs = -1,
            class_weight = 'balanced',
            random_state = 94
)
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_test)
y_proba = final_model.predict_proba(X_test)[: , 1]

report = classification_report(y_test , y_pred)

cm_CL = confusion_matrix(y_test , y_pred)

plotConfusionMatrix(cm_CL)

print(report)
```

# ROC-CURVE

```python
fpr , tpr , thr = roc_curve(y_test , y_proba)
roc_auc = auc(fpr, tpr)

J = tpr - fpr
ix = np.argmax(J)
best_threshold = thr[ix]

plt.figure(figsize=(6,5))
plt.plot(fpr , tpr , label = 'AUC = %.3f' % roc_auc)
plt.scatter(fpr[ix], tpr[ix], marker='o', label='Best Threshold')
plt.plot([ 1 , 0] , [1 ,0]   , '--')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.title('Reciver Operational Characteristic Curve(ROC)')
plt.legend()
plt.show()
```

# Precision Recall Curve

```python
prec , recall , thresholds =  precision_recall_curve(y_test , y_proba)
pr_area = average_precision_score(y_test , y_proba)

f1_scores = 2 * (prec * recall) / (prec + recall)
best_idx = np.argmax(f1_scores)
best_threshold_pr = thresholds[best_idx] 

plt.figure(figsize=(6,5))
plt.plot(prec , recall , label = 'Area Under PR-curve = %.3f' % pr_area)
plt.scatter(prec[best_idx] , recall[best_idx] , marker='o' , label='Best Threshold')
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

# Threshold Optimisation

## PR-Curve basesed threshold optimization

```python
# recalculation predictions with best threshold
y_pred_pr = (y_proba >= best_threshold_pr).astype('int')
cm_pr = confusion_matrix(y_test , y_pred_pr)

print('Best Threshold for based on PR-Curve:' , best_threshold_pr)
plotConfusionMatrix(cm_pr)
print(classification_report(y_test, y_pred_pr))
```

## Cost-based threshold optimization

```python
# 1000$ for missing Fraud
# 10$ for false alarm

cost_fn = 1000
cost_fp = 10

all_thresholds = np.linspace(0 , 1 , 100)

costs = []

for t in all_thresholds:
    y_pred_c = (y_proba >= t).astype('int')

    tn , fp , fn , tp = confusion_matrix(y_test , y_pred_c).ravel()

    total_cost = (cost_fn * fn) + (cost_fp * fp)

    costs.append(total_cost)

best_threshold_cost = all_thresholds[np.argmin(costs)]
print("Best Threshold for cost-based:", best_threshold_cost)

y_pred_c = (y_proba >= best_threshold_cost).astype('int')

cm_cost = confusion_matrix(y_test , y_pred_c)
plotConfusionMatrix(cm_cost)
print(classification_report(y_test , y_pred_c))
    
```
